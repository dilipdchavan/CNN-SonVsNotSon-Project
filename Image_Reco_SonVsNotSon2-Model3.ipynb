```python
import keras
keras.__version__
from keras.utils.np_utils import to_categorical
```

```python
from keras.models import load_model

model = load_model('son_and_notson_model_3.h5')
model.summary()  # As a reminder.
```

```python
#Model 3 : Test Image 1
```

```python
img_path = 'C:\\DCHAVAN\\DSTI-Masters\\DilipPersonalStudy\\DeepLearning\\ImageReco_Face\\Test_Images-Models\\Test-Image1.jpg'

# We preprocess the image into a 4D tensor
from keras.preprocessing import image
import numpy as np

img = image.load_img(img_path, target_size=(150, 150))
img_tensor = image.img_to_array(img)
img_tensor = np.expand_dims(img_tensor, axis=0)
# Remember that the model was trained on inputs
# that were preprocessed in the following way:
img_tensor /= 255.

# Its shape is (1, 150, 150, 3)
print(img_tensor.shape)
```

```python
import matplotlib.pyplot as plt

plt.imshow(img_tensor[0])
plt.show()
```

```python
prediction1 = model.predict(img_tensor)
```

```python
print(prediction1)
```

```python
prediction1=np.round(prediction1,decimals=2)
labels1=['1-Son' if value>0.5 else '0-NotSon' for value in prediction1]
```

```python
print(prediction1)
```

```python
print(labels1)
```
